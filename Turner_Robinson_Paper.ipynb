```python
import numpy as np
import matplotlib.pyplot as plt

def Generate_K(num_points, radius = 1, noise_rho=0.1):
    # Generate random angles
    angles = np.random.uniform(0, 2*np.pi, num_points)
    
    # Generate points on the circle without noise
    x = radius * np.cos(angles)
    y = radius * np.sin(angles)
    
    # Add Gaussian noise to the points
    noise_x = np.random.normal(0, noise_rho, num_points)
    noise_y = np.random.normal(0, noise_rho, num_points)
    
    # Add noise to points
    x_noisy = x + noise_x
    y_noisy = y + noise_y
    
    return x_noisy, y_noisy

num_points = 50
radius = 1.0
noise_rho = 0.05
x, y = Generate_K(num_points, radius, noise_rho)

# Plotting the point cloud
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5, alpha=0.5)
plt.title(f"K point cloud | (rho={noise_rho})")
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

```python
def Generate_M(num_points, Inner_Radius = 0.9, Outer_Radius = 1.1, noise_rho=0.1):
    # Generate random angles
    angles = np.random.uniform(0, 2*np.pi, num_points)
    
    radii = np.sqrt(np.random.uniform(Inner_Radius **2, Outer_Radius **2, num_points))
    # Generate points on the circle without noise
    x = radii * np.cos(angles)
    y = radii * np.sin(angles)
    
    # Add Gaussian noise to the points
    noise_x = np.random.normal(0, noise_rho, num_points)
    noise_y = np.random.normal(0, noise_rho, num_points)
    
    # Add noise to points
    x_noisy = x + noise_x
    y_noisy = y + noise_y
    
    return x_noisy, y_noisy

num_points = 50
r1 = 0.9
r2 = 1.1
noise_rho = 0.05
x, y = Generate_M(num_points, r1, r2, noise_rho)

# Plotting the point cloud
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=5, alpha=0.5)
plt.title(f"M point cloud | (rho={noise_rho})")
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```
